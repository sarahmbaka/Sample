```python
import pandas as pd
import requests
import json

#data dictionaries
data_directory = "../data/"
data_output_directory = data_directory + "notebook_outputs/" 
data_flatfile_directory = data_directory + "flatfile/"

test_data = pd.read_csv(data_flatfile_directory+"all_features_test_1579156846.csv")
```

```python
test_data.head()
```

```python
#Function to assist in making requests and getting response in SuperML
def sfl_request(df, superml_api_url):
    for column in df.columns.tolist():
        df[column] = df[column].astype(str)
    df = df.to_dict(orient='records')
    df = dict(req_objects=df)
    df = dict(post_data=df)
    response = requests.post(superml_api_url,
                    data=json.dumps(df), verify=False, headers={'content-type':'application/json'})
    api_results = response.json().get('result').get('predictions')    
    return pd.DataFrame(api_results)
```

```python
dropped_columns = ['status_by_margins_60', 'order_id', 'order_value.group','total_arrears', 'age_of_loan', 'margin','status_by_margins_30']

test_df = test_data.drop(dropped_columns, axis=1)
```

```python
test_predictions=sfl_request(test_df,'https://app.superml.io/api/v1/prediction/ml/3128ea5addd37f2fd58755ba945da87683f862a099dd2de0b51a')
```
